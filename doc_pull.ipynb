```python
from sec_api import QueryApi
import json

"""
ticker: company ticker name
start_date / end_date need to be in format "year-month-day". Write '1' as '01' for month and day. Ex: "2021-01-01"
form_type: "10-Q", "10-K", "8-K"
"""
def pull_forms(ticker, start_date, end_date, form_type):
    queryApi = QueryApi(SEC_API_KEY)

    embedded_query = f"ticker:{ticker} AND filedAt:[{start_date} TO {end_date}] AND formType:\"{form_type}\""

    query = {
      "query": embedded_query,
      "from": "0",
      "size": "10",
      "sort": [{ "filedAt": { "order": "desc" } }]
    }

    response = queryApi.get_filings(query)

    return response["filings"]
    
```

```python
def pull_section(ticker, start_date, end_date, form_type, section):
    toReturn = []
    data = pull_forms(ticker, start_date, end_date, form_type)
    for i in range(len(data)):
        link = data[i]["linkToFilingDetails"]
        filing = [data[i]['periodOfReport'], link, extractorApi.get_section(link, section, "text")]
        toReturn.append(filing)
    return toReturn
```

```python
%pip install langchain-community langchain-openai ipywidgets
from langchain_community.chat_models import ChatOpenAI
from langchain_core.prompts.chat import ChatPromptTemplate, HumanMessagePromptTemplate, SystemMessagePromptTemplate
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from IPython.display import display
from ipywidgets import widgets
import json
```

```python
chatbot = ChatOpenAI(temperature=0, openai_api_key=OPENAI_API_KEY, model='gpt-3.5-turbo')
summarizer = ChatOpenAI(temperature=0, openai_api_key=OPENAI_API_KEY, model='gpt-3.5-turbo')
doc_puller = ChatOpenAI(temperature=0, openai_api_key=OPENAI_API_KEY, model='gpt-3.5-turbo')
```

```python
chatbot_prompt = open("chatbot_prompt.txt", 'r').read()
summarizer_prompt = open("summarizer_bot.txt", 'r').read()
doc_puller_prompt = open("doc_puller_bot_prompt.txt", 'r').read()

# Initialize conversation history
chatbot_conversation_history = [
    SystemMessage(content=chatbot_prompt)
]

summarizer_conversation_history = [
    SystemMessage(content=summarizer_prompt)
]

doc_puller_conversation_history = [
    SystemMessage(content=doc_puller_prompt)
]

print("Chatbot initialized. Type 'exit' to end the conversation.")

text_input = widgets.Text(
    description='You:',
    continuous_update=False
)

def on_submit(change):
    user_input = change['new']
    
    # Check if user wants to exit
    if user_input.lower() == 'exit':
        print("Chatbot: Goodbye!")
        return
    

    doc_puller_conversation_history.append(HumanMessage(content=user_input))
    dp_response = doc_puller(doc_puller_conversation_history)
    # see whether to pop last element or not
    doc_puller_conversation_history.pop() 

    chatbot_conversation_history.append(HumanMessage(content=user_input))

    print(dp_response)
    
    
    # if chatbot response is empty -> skip
    if dp_response.content == "None":
        c_response = chatbot(chatbot_conversation_history)
        print(c_response.content)

        chatbot_conversation_history.append(AIMessage(content=c_response.content))
    

    elif isinstance(dp_response.content, str):
        
        # if chatbot response is JSON -> extract
        try:
            content_json = json.loads(dp_response.content)
            # Extract parameters from the JSON
            ticker = content_json.get('ticker')
            start_date = content_json.get('start_date')
            end_date = content_json.get('end_date')
            form_type = content_json.get('form')
            sections = content_json.get('sections')

            print("JSON:", content_json)

            # Call the pull_section() function with the extracted parameters

            q_context = []
            for section in sections:
                texts = pull_section(ticker, start_date, end_date, form_type, section)
                for t in texts:
                    q_context.append(t[2])
            

            q_context_text = ""
            for t in q_context:
                q_context_text += t

            summarized_responses = ""
            
            for i in range(0, len(q_context_text), 55000):
                chunk = q_context_text[i:i+55000]
                summarizer_conversation_history.append(HumanMessage(content=f"query: {user_input}. chunk: {chunk}"))
                s_response = summarizer(summarizer_conversation_history)
                summarizer_conversation_history.pop()
                summarized_responses += f"Context Chunk: {s_response}\n"

            

            chatbot_conversation_history.append(SystemMessage(content=f"Here is some context to use when answering the previous user message. Context: {summarized_responses}"))
            c_response = chatbot(chatbot_conversation_history)

            print(c_response)
            chatbot_conversation_history.append(AIMessage(content=c_response.content))


        # if chatbot response is text -> send to user and add to chatbot history
        except json.JSONDecodeError:
            print(dp_response.content)
            chatbot_conversation_history.append(AIMessage(content=dp_response.content))


text_input.observe(on_submit, names='value')
display(text_input)
```
